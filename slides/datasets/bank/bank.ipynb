```python
import pandas as pd
import zipfile
from io import StringIO
import requests
from io import BytesIO

df = pd.read_csv("./datasets/bank/bank-additional.csv", sep=';')
df
```

```python
from sklearn.model_selection import train_test_split
from flaml import AutoML

# Split the data into features (X) and target (y)
X = df.drop(columns=['y'])
y = df['y']

# Perform train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize AutoML
automl = AutoML()

# Define settings for AutoML
automl_settings = {
    "time_budget": 30,  # Time budget in seconds
    "max_iter": 30,  # Max number of iterations
    "metric": 'accuracy',  # Evaluation metric
    "task": 'classification',  # Task type
    "log_file_name": "automl.log",  # Log file
}

# Run AutoML
automl.fit(X_train, y_train, **automl_settings)

# Evaluate on the test set
print("Best ML model:", automl.best_estimator)
print("Accuracy on test set:", automl.score(X_test, y_test))
```
